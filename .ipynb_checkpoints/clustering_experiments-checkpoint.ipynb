```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_file = "../countries5.csv"
data = pd.read_csv(data_file)
data_new = data.drop(["IVR"], axis = 1)
data_new.set_index("Country", inplace = True)
```

```python
data_new
```

```python
Karray = [2,3,7,10]
total_output = []
for x in Karray:
    output = []
    for i in range(10):
        kmeans = KMeans(n_clusters= x, init = 'random',n_init = 50).fit(data_new)
        output.append(kmeans.inertia_)
    print("k = ", x, "with SSR= ", output, "\n")
    total_output.append(output)
```

```python

        kmeans1 = KMeans(n_clusters= 2, init = 'random',n_init = 50).fit(data_new)
        print(kmeans1.inertia_)
        kmeans = KMeans(n_clusters= 2).fit(data_new)
        print(kmeans.inertia_)
```

```python
Sum_of_squared_distances = []
K = range(1, 15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_new)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show
```

```python
elbow_kmeans = KMeans(n_clusters= 5).fit(data_new)
print(elbow_kmeans.inertia_)
```

```python
clusters = []
for i in range(5):
    mask = elbow_kmeans.labels_ == i
    cluster = data[mask]
    cluster = cluster.loc[:,"Country"]
    print("The ",i,"th cluster is: \n", cluster, "\n")
    print("with centroid: ", elbow_kmeans.cluster_centers_[i])
    clusters.append(cluster)
```

```python
label = data.loc[:,"Country"]
def llf(id):
    if id< 81:
        return label[id]
    else:
        return ""
    
from matplotlib.pyplot import figure
from scipy.cluster.hierarchy import dendrogram



def plot_dendrogram(model):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    distance = np.arange(model.children_.shape[0])
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, distance, counts]
    ).astype(float)
    
    fig = plt.figure()
    fig.set_size_inches(15,15)
    ax = fig.add_subplot(1,1,1)
    dendrogram(linkage_matrix, leaf_label_func = llf)
    ax.tick_params(axis = 'x', which = 'major', labelsize = 11)
    ax.tick_params(axis = 'y', which = 'major', labelsize = .5)
    plt.savefig('t1.png')



plot_dendrogram(clustering)
```
